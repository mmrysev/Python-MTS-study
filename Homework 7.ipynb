2)Постройте графики функции на одном полотне и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5. Убедитесь, что по оси X значения изменяются в диапазоне [-10, 10].

```python
x=range(-10,11,1)
y1 = [i**3 for i in x]
y2 = [2*i**3 for i in x]
```

```python
plt.plot(x,y1,'b',label='$y=x^3$')
plt.plot(x,y2,'r',label='$y=2*x^3$')
plt.fill_between(x, y1, y2, color='y', alpha=0.5, label=u'область')
plt.title(u'Графики', family="verdana")
plt.xlabel(u'ось Х', family='verdana')
plt.ylabel(u'ось У', family='verdana', rotation='horizontal')
plt.legend(loc = 2)
plt.show()
```

Давайте посмотрим, насколько сильно различаются между собой объекты классов Malignan и Benign по паре признаков 'mean radius' и 'mean texture', а также по паре признаков 'mean perimeter' и 'mean area'. Для этого нужно изобразить объекты этих классов разными цветами в координатах соответсвующих пар признаков (для работы с цевтом подойдет ListedColormap)  и проанализировать полученные изображения. 

На одном полотне отобразите объекты в координатах 'mean radius' и 'mean texture', а на другом - в координатах 'mean perimeter' и 'mean area'. Расположите оба полотна на одном изображении горизонтально (используйте plt.subplot). Как вы считаете, сложно ли будет отличить объекты друг от другах по рассматриваемым парам признаков?

```python
from matplotlib.colors import ListedColormap
```

```python
colors = ListedColormap(['yellow', 'violet'])
```

```python
plt.figure(figsize=(12,4))

plt.subplot(121)

plt.scatter(df.loc[:,'mean radius'], df.loc[:,'mean perimeter'], c = df.target, 
            cmap = colors, alpha=0.5,marker='x')
plt.xlabel('mean radius')
plt.ylabel('mean perimeter')

plt.subplot(122)

plt.scatter(df.loc[:,'mean perimeter'], df.loc[:,'mean area'], c = df.target, 
            cmap = colors,alpha=0.3)
plt.xlabel('mean perimeter')
plt.ylabel('mean area')
plt.legend(loc=2)

plt.show()

dims = [('mean radius',1),('mean area',100), ('mean perimeter',10)]

plt.figure(figsize=(12,5))

i=1
for dim, s in dims:
    plt.subplot(130+i)

    plt.title(dim)
    plt.hist(df.loc[df.target == 0,dim], color='g', alpha=0.5, label=data.target_names[0])
    plt.hist(df.loc[df.target == 1,dim], color='b', alpha=0.5, label=data.target_names[1])
    
    p = df.groupby(df[dim].apply(lambda x: int(x/s)))[dim,'target'].mean()
    plt.plot(p.loc[:,dim],p.loc[:,'target']*100,'r',label=u'вероятность\nbenign')
    
    plt.legend(loc='upper right')
    
    i+=1
```

Между параметрами практически линейная зависимость, классификацию можно проводить на одном параметре из них. Для повышения точности классификации нужно анализировать связь с другими параметры выборки.
